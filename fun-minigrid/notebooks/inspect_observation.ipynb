# Inspect MiniGrid Observation

MiniGrid DoorKey 환경의 observation shape과 action space 크기를 확인한다.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT))

from src.envs.make_env import make_env
from src.envs.preprocess import preprocess_obs
```

```python
env = make_env(seed=42)
obs, info = env.reset(seed=42)

print("observation shape:", obs.shape)
print("observation dtype:", obs.dtype)
print("action space:", env.action_space)
print("action space size:", env.action_space.n)

env.close()
```

```python
env = make_env(seed=42)
obs, info = env.reset(seed=42)
processed_obs = preprocess_obs(obs)

print("raw shape:", obs.shape)
print("raw dtype:", obs.dtype)
print("processed shape:", processed_obs.shape)
print("processed dtype:", processed_obs.dtype)
print("processed min:", processed_obs.min())
print("processed max:", processed_obs.max())

env.close()
```
